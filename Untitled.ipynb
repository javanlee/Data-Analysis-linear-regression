```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
slope_1, intercept_1, r_1, p_1, se_1 = linregress(df['Year'],df['CSIRO Adjusted Sea Level'])
df_2000=df[df['Year']>=2000]
slope_2, intercept_2, r_2, p_2, se_2 = linregress(df_2000['Year'],df_2000['CSIRO Adjusted Sea Level'])
#
fig, axes = plt.subplots()
#df.plot('Year',y='CSIRO Adjusted Sea Level',kind='scatter')
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])
plt.axline(xy1=(0, intercept_1), slope=slope_1, label=f'$y = {slope_1:.8f}x {intercept_1:+.8f}$',color='k')
#
plt.axline(xy1=(0, intercept_2), slope=slope_2, label=f'$y = {slope_2:.8f}x {intercept_2:+.8f}$',color='r')
axes.set_xlim([1875, 2050])
axes.set_ylim([-5, 16])
axes.xaxis.set_ticks([1850.0, 1875.0, 1900.0, 1925.0, 1950.0, 1975.0, 2000.0, 2025.0, 2050.0, 2075.0])
axes.set(xlabel='Year', ylabel='Sea Level (inches)')
axes.set_title('Rise in Sea Level')
fig.savefig('teste.png')
```
